```python
#Creating a function to reposition columns
def movecol(df, cols_to_move=[], ref_col='', place='After'):
    
    #cols_to_move can be one or more columns we wish to move
    #ref_col is the column after or before which we want to reposition the column(s)
    
    cols = df.columns.tolist()
    if place == 'After':
        seg1 = cols[:list(cols).index(ref_col) + 1]
        seg2 = cols_to_move
    if place == 'Before':
        seg1 = cols[:list(cols).index(ref_col)]
        seg2 = cols_to_move + [ref_col]
    
    seg1 = [i for i in seg1 if i not in seg2]
    seg3 = [i for i in cols if i not in seg1 + seg2]
    
    return(df[seg1 + seg2 + seg3])
```

```python
netflix_content_data = pd.DataFrame(columns = ['Country', 'Description', 'Total Library Size', 'No. of TV Shows', 'No. of Movies','Cost Per Month - Basic ($)', 'Cost Per Month - Standard ($)','Cost Per Month - Premium ($)'])

#Largest Library
max_lib = netflix[netflix['Total Library Size'] == netflix['Total Library Size'].max()]
max_lib.insert(0, "Description", "Largest Content Library", True)

#Smallest Library
min_lib = netflix[netflix['Total Library Size'] == netflix['Total Library Size'].min()]
min_lib.insert(0, "Description", "Smallest Content Library", True)

#Largest TV Library
max_tv = netflix[netflix['No. of TV Shows'] == netflix['No. of TV Shows'].max()]
max_tv.insert(0, "Description", "Largest Library of TV Shows", True)

#Smallest TV Library
min_tv = netflix[netflix['No. of TV Shows'] == netflix['No. of TV Shows'].min()]
min_tv.insert(0, "Description", "Smallest Library of TV Shows", True)

#Largest Movie Library
max_mov = netflix[netflix['No. of Movies'] == netflix['No. of Movies'].max()]
max_mov.insert(0, "Description", "Largest Library of Movies", True)

#Smallest Movie Library
min_mov = netflix[netflix['No. of Movies'] == netflix['No. of Movies'].min()]
min_mov.insert(0, "Description", "Smallest Library of Movies", True)

#Adding columns to empty dataframe
netflix_content_data = pd.concat([netflix_content_data, max_lib, min_lib, max_tv, min_tv, max_mov, min_mov], axis=0)

#Removing extra columns
netflix_content_data = netflix_content_data[['Country', 'Description', 'Total Library Size', 'No. of TV Shows', 'No. of Movies']]


netflix_content_data
```

```python
netflix_subscription_data = pd.DataFrame(columns = ['Country', 'Description', 'Total Library Size', 'No. of TV Shows', 'No. of Movies','Cost Per Month - Basic ($)', 'Cost Per Month - Standard ($)','Cost Per Month - Premium ($)'])

#Highest basic cost per month
max_basic = netflix[netflix['Cost Per Month - Basic ($)'] == netflix['Cost Per Month - Basic ($)'].max()]
max_basic.insert(0, "Description", "Highest Subsription Costs", True)

#Smallest basic cost per month
min_basic = netflix[netflix['Cost Per Month - Basic ($)'] == netflix['Cost Per Month - Basic ($)'].min()]
min_basic.insert(0, "Description", "Lowest Subsription Costs", True)

#Highest standard cost per month
max_standard = netflix[netflix['Cost Per Month - Standard ($)'] == netflix['Cost Per Month - Standard ($)'].max()]
max_standard.insert(0, "Description", "Highest Standard Subsription", True)

#Smallest standard cost per month
min_standard = netflix[netflix['Cost Per Month - Standard ($)'] == netflix['Cost Per Month - Standard ($)'].min()]
min_standard.insert(0, "Description", "Lowest Standard Subsription", True)

#Highest premium cost per month
max_premium = netflix[netflix['Cost Per Month - Premium ($)'] == netflix['Cost Per Month - Premium ($)'].max()]
max_premium.insert(0, "Description", "Highest Premium Subsription", True)

#Smallest premium cost per month
min_premium = netflix[netflix['Cost Per Month - Premium ($)'] == netflix['Cost Per Month - Premium ($)'].min()]
min_premium.insert(0, "Description", "Lowest Premium Subsription", True)

#Adding columns to empty dataframe
netflix_subscription_data = pd.concat([netflix_subscription_data, max_basic, min_basic, max_standard, min_standard, max_premium, min_premium], axis=0)


netflix_subscription_data
```
